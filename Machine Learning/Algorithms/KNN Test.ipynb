```python
import pandas as pd
import numpy as np
from collections import Counter
from sklearn.metrics import f1_score
```

```python
df = pd.read_csv('../../data/heart.csv')
```

```python
df.describe()
```

```python
# Calculate mean
mean = np.array(df.iloc[:, [0,3,4,7,9]].mean())
# Calculate std
std = np.array(df.iloc[:, [0,3,4,7,9]].std())
```

```python
df.iloc[:, [0,3,4,7,9]]
```

```python
# Standardizing
df.iloc[:, [0,3,4,7,9]] = (df.iloc[:, [0,3,4,7,9]] - mean) / std
df.iloc[:, [0,3,4,7,9]]
```

```python
y_train = df.pop('target').values
x_train = df.values
```

```python
def KNN(x_test, X, Y, k):
    prediction = []
    for x in x_test:
        instance_array = np.tile(x, (X.shape[0],1))
        distance_array = np.sum(np.square(instance_array - X),axis = 1)
        sorted_indices = np.argsort(distance_array)[0:k]
        label_indices = Y[sorted_indices]
        counted_dict = Counter(label_indices).most_common()
        label = counted_dict[0][0]
        prediction.append(label)
    return prediction
        

predictions = KNN(x_train, x_train, y_train, 20)                        
```

```python
f1_score(y_train, predictions, average = 'macro')
```
