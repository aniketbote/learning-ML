```python
import pandas as pd
```

```python
df = pd.read_csv('../Algorithms/LogMetric_k_p_tuned.csv')
```

```python
df.sort_values(by=['score','scaling','scaler'], ascending = False, inplace = True)
df['p'] = df['p'].apply(lambda x : x - 0.1)
df.to_csv('LogMetric_tuned.csv', index= False)
```

```python
above8 = df[df['score'] > 0.85]
above8.loc[:,['ca', 'cp', 'exang', 'fbs', 'restecg',
       'sex', 'slope', 'thal']].apply(pd.value_counts)
```

```python
above8.groupby(['scaler','ca', 'cp', 'exang', 'fbs', 'restecg','sex', 'slope', 'thal']).mean()
```

```python
pd.set_option('display.max_rows', 2000)
df.groupby(['scaler','ca', 'cp', 'exang', 'fbs', 'restecg','sex', 'slope', 'thal']).mean().sort_index()
```

```python
df.groupby(['scaler']).mean().sort_index().loc[:,['score']]
```

```python
groups = df[df['scaler'] == 'robust'].groupby(['ca', 'cp', 'exang', 'fbs', 'restecg','sex', 'slope', 'thal']).mean().sort_index()['score']
groups
```

```python
robust_group = list(groups)
```

```python
plot_df = pd.DataFrame()
plot_df['robust'] = robust_group
plot_df['robust'].plot()
```
