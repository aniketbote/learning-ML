```python
def scatter_plot(data, y, x1, x2 = None, x3 = None, x4 = None):
    if x4 != None and len(set(data[x4])) > 6 :
        print('X4 should not have more than 6 distinct values.')
        x4 = None
    plt.show(sns.relplot(x = x1, y = y, hue = x2, size = x3, style = x4, data = data, palette = "inferno"))
```

```python
def line_plot(data, y, x1, x2 = None, x3 = None, x4 = None, estimator = None,sort = True):
    if x4 != None and len(set(data[x4])) > 6 :
        print('X4 should not have more than 6 distinct values.')
        x4 = None
    palette = None
    if x2 != None:
        m = len(set(data[x2]))
        palette = sns.cubehelix_palette(light = 0.5, n_colors = m)

    plt.show(sns.relplot(x = x1, y = y, hue = x2, size = x3, style = x4, data = data, kind = 'line', palette = palette, estimator = estimator, sort = sort, markers = True))
    
```

```python
def cat_plot(data, y, x1, x2 = None, kind = 'strip'):
    sns.catplot(kind = kind, data = data, x = x1, y = y, hue = x2)
```

```python
def dist_plot(x, kde = True, hist = True):
    sns.distplot(x, kde = kde, hist = hist)
```

```python
def joint_plot(data,x,y,kind = 'scatter'):
    sns.jointplot(data = data, x = x, y = y, kind = kind)
```

```python
def pair_plot(data, diag = 'hist', offdiag = 'scatter'):
    if diag == 'hist' and offdiag == 'scatter':
        sns.pairplot(data)
    elif diag == 'hist' and offdiag == 'kde':
        g = sns.PairGrid(data)
        g.map_diag(plt.hist)
        g.map_offdiag(sns.kdeplot)
    elif diag == 'kde'and offdiag == 'line':
        g = sns.PairGrid(data)
        g.map_diag(sns.kdeplot)
        g.map_offdiag(sns.lineplot)
```
